```python
%%cython -a

cdef class A(object):
    cdef float x, y
    
    def __cinit__(self, float x, float y):
        self.x = x
        self.y = y
        
    cdef void test(self):
        self.x += 1
        
    @staticmethod
    cdef float t(float a):
        return a + a
    
def b(float a):
    return A.t(a)

cdef class B(A):
    
    cdef void test(self):
        self.x += 2
```

```python
a = A(5, 6)
```

```python
b = B(5, 7)
```

```python
b(3)
```

```python
%%cython -a

cdef class A:
    cdef:
        float x
        void (*func)(A)
    
    def __cinit__(self):
        pass
    
    def getX(self):
        return self.x
    
cdef void a(A obj):
    obj.x += 1
```

```python
%%cython -a

from macrospin cimport field, torque, energy
from macrospin.types cimport *

cdef class MyKernel(Kernel):
    
    cdef float3 field(self, float t, float3 m):
        return self.hext + field.demagnetization(m, self.Nd)
    
    cdef float3 torque(self, float t, float3 m):
        float3 heff = self.field(t, m)
        return torque.landau_lifshitz(m, heff, self.alpha)
    
    cdef float energy(self, float t, float3 m):
        return energy.zeeman(self.field(t, m))
```
