The material yttrium iron garent (YIG) is a ferrimagnet, with two Fe$^{3+}$ sub-lattices composing its magnetic structure. This notebook will summarize the magnetic behavior of YIG as a function of temperature to allow quantitative predictions to be made for the material.

## Saturation magnetization ##

The saturation magnetization of YIG increases monotonically with decreasing temperature. Using the two sublattice model for ferrimagnetics, the molecular field equations for YIG are as follows, based on [Anderson, 1964](http://dx.doi.org/10.1103/PhysRev.134.A1581).

$$ \sigma = \sigma_d - \sigma_a $$
$$ \sigma_a = 75.70 \times B_{5/2}\left(\frac{3.359 \times 10^{-4}}{T}\right) H_a $$
$$ \sigma_d = 113.55 \times B_{5/2}\left(\frac{3.359 \times 10^{-4}}{T}\right) H_d $$
$$ H_a = \lambda_{aa} \sigma_a - \lambda_{ad} \sigma_d $$
$$ H_d = -\lambda_{ad} \sigma_a - \lambda_{dd} \sigma_d $$

In the above equations, $\sigma$ is the total magnetization, $\sigma_a$ is the magnetization of the octahedral sublattice, and $\sigma_d$ is the magnetization of the tetrahedral sublattice, in terms of emu/g.

This is illustrated in the experimental data by [Anderson, 1964](http://dx.doi.org/10.1103/PhysRev.134.A1581).
